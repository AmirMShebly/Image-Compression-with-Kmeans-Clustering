```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from matplotlib.image import imread
```

```python
img = imread('pngtree-powder-smoke-colorful-background-image_2164096.jpg')
```

```python
print(img.max())
print(img.min())
```

```python
img = img / 255
```

```python
h, w, c = img.shape
X_img = np.reshape(img, (h * w, c))
```

```python
plt.imshow(img)
```

```python
K = 16
max_iters = 10 
```

```python
kmeans = KMeans(n_clusters=K, max_iter=max_iters)
kmeans.fit(X_img)
centroids = kmeans.cluster_centers_
idx = kmeans.labels_
```

```python
X_recovered = centroids[idx, :]
X_recovered = np.reshape(X_recovered, (h, w, c))
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 8))
ax[0].imshow(img)
ax[0].set_title('Original')
ax[0].axis('off')

ax[1].imshow(X_recovered)
ax[1].set_title(f'Compressed with {K} colors')
ax[1].axis('off')

plt.show()
```
